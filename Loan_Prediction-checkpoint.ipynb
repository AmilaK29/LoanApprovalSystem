# Data Preprocessing

There are two steps 
       1. Encoding
       2. Data cleaning

```python
X = training_data.drop(columns = ["Loan_Status"])
Y = training_data[["Loan_Status"]]
X.columns.tolist()
```

```python
for i in X:
    if(i in cat_cols):
        print(i)
        X[i].fillna(value = X[i].mode()[0],inplace = True)
    else:
        X[i].fillna(value = X[i].mean(),inplace = True)
```

```python
X.info()
```

```python
X = pd.get_dummies(X,drop_first = True)
X.head()
```

```python
Y = pd.get_dummies(Y,drop_first = True)
Y.head()
```
